```python
# Copying a List
## 단순히 = 연산자로 대입할 경우, 동일한 메모리 주소를 참조하는 식으로 변함
## 파이썬의 모든 변수는 해당 메모리 주소를 참조하는 이름이다.
a = [1, 2, 3]
b = a 
print(id(a), id(b))

## Shallow Copy : 얕은 복사
###     :: Mutable in Mutable인 경우, 복사가 되지 않음. 동일한 요소를 참조하고 있기에 한쪽을 바꾸면 다른쪽도 변화.
### [:]
b = a[:]
print(id(a), id(b))

### copy module
import copy
b = copy.copy(a)
print(id(a), id(b))

## Deep Copy
b = copy.deepcopy(a)
print(id(a), id(b))
```

```python
# Class 
## 파이썬 메서드의 첫번째 매개변수는 관례적으로 self를 사용한다.
## 메소드를 호출할 때 호출한 객체 자신이 전달되기 때문에 명시적으로 self를 넣어주며, 메소드 호출 시에는 생략한다.
### Constructor(생성자)
#### : __init__ 이라는 이름의 함수
class FourCalc:
    def __init__(self, first, second):
        self.first = first
        self.second = second

    def setdata(self, first, second):
        self.first = first 
        self.second = second 
    
    def add(self):
        result = self.first + self.second
        return result
    def sub(self):
        result = self.first - self.second
        return result
    def mul(self):
        result = self.first * self.second
        return result
    def div(self):
        result = self.first / self.second
        return result

### Inheritance(상속)
#### class InheritClass(ToInherit):
class MoreFourCal(FourCalc):
    def pow(self):
        result = self.first ** self.second
        return result

### Method Override(함수 재정의)
class SafeFourCal(FourCalc):
    def div(self):
        if self.second == 0: return 0
        else: 
            return self.first / self.second



## Class Variables (클래스 변수) 
## : 클래스로 접근 시, 해당 클래스로 만든 모든 객체들이 공유하는 변수
class Family:
    lastname = 'Kwon'

a = Family(); b = Family()
print(a.lastname, b.lastname)
Family.lastname = 'Kim'
print(a.lastname, b.lastname)
a.lastname = 'Park' # Family.lastname이 바뀌는 것이 아닌, a객체에 lastname이라는 새로운 객체 변수가 초기화되어 할당된다.
print(Family.lastname, a.lastname)
```

```python
# Enumerate()
## 파이썬에서 iterable의 인덱스와 원소를 동시에 접근하여 루프를 돌릴 수 있게 해주는 함수
## 인덱스와 원소를 묶어서 튜플로 만든다.

for entry in enumerate(['a', 'b', 'c']):
    print(entry)
print()
# Unpacking
for index, value in enumerate(['A', 'B', 'C']):
    print(index, value)
print()

# Change Start index
for i, v in enumerate(['A', 'B', 'C'], start=101):
    print(i, v)
print()

## 응용
matrix = [['A', 'B', 'C'], ['D', 'E', 'F'], ['G', 'H', 'I']]

for r, row in enumerate(matrix):
    for c, letter in enumerate(row):
        print (r, c, letter)
```

```python
## sorted() 함수의 활용
N = int(input())
sells = dict()

for _ in range(N) :
    book = input()
    if book in sells :
        sells[book] += 1
    else :
        sells[book] = 1

# sorted() => dictionary key값 기준 default 오름차순 정렬, reverse=True를 통해 내림차순 정렬 가능
# key=lambda (x: (x[0], -x[1])) => 특정 키 값을 중심으로 정렬
# sells = sorted(sells.items())
print(sorted(sells.items()))
print(sorted(sells.items(),reverse=True))
print(sorted(sells.items(), key=lambda x: x[0]))
print(sorted(sells.items(), key=lambda x: -x[1]))
print(sorted(sells, reverse=True))
# print(sorted(sells, key=lambda x: x[1], reverse=True)[0][0])
# print(sorted(sells, key=lambda x: x[1]))
```
